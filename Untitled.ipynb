### In this tutorial we demonstrate how to use matplotlib to generate visuals

```python
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn
plt.style.use('ggplot')
```

```python
# Creating a simple plot
x = np.arange(7)
plt.plot(x, 'g+--', lw=0.8)
# Here, the g is for color green, * is for the markers, and -- is for the line style
plt.show()
```

### The other marker options include
'.',   'o',   '<',   '>', 'v', '^', '1', '2', '3', '4', 's', 'p', '*', 'h', 'H', '+', 'D', 'd', '|', '_'

### Changing the ticks

```python
y = np.linspace(-6, 4, 1000)
z = np.sin(y)
plt.plot(y, z)

# Here we are substituting the original xticks with a selected list of our choice
plt.xticks([-6, -5, -4, -3, -2, -1, 0, 1, 2, 3, 4], ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h','i', 'j', 'k'])

# Here we are selecting the yticks we need for the plot
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.show()
```

# Styles
A style dictates things such as marker size, colors, and fonts. 

There are many built-in styles in Matplotlib. This notebook has been set to make use of the 'ggplot' style, but there are several options to chose from:

To print the several options we make use of the following code:

`print(plt.style.available)`

### Styling is applied globally to the entire notebook.
However, if we need to change styling locally we proceed as follows:


```python
# We make use of the block of code:

with plt.style.context('Solarize_Light2'):
    data = [0,1,2,3,4,5]
    plt.plot(np.sin(data), 'g.--')
    plt.title('plot using "Solarize_Light2" style')
    plt.show()
```

```python
def simple_plot(style):
    # This is a function that takes in a style as a parameter and applies that style in the plot
    with plt.style.context(style):
        x = np.arange(4)
        plt.plot(x)
        plt.title(style)
        plt.show()
```

## Demonstrate plots using the different styles 
The following code will create the same plot using different styles, so as to see the differences in the styles.

```python
# Create a list of all styles available:
lst = plt.style.available

# Create a figure object
fig = plt.figure(figsize=(20, 40))

for i, j in enumerate(lst):
    
    # Select a style to use for each plot:
    with plt.style.context(j):
        
        # Create each subplot:
        ax = fig.add_subplot(7, 4, i+1)
        
        # Plot the data
        x = np.arange(4)
        plt.plot(x**2)
        plt.title(j)
plt.show()
```
